```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# Checking the data displayed in the dataframe
school_data_complete
```

```python
#Calculate the total number of students
Total_Students = school_data_complete["Student ID"].count()
Total_Students 
```

```python
# Get a list of all of our columns for easy reference
school_data_complete.columns
```

```python

#  Create a new df for "Average math score,average reading score,%math score,%reading score, " with the columns. 

#calculate Average math score
average_math_score = school_data_complete['math_score'].mean()
print(average_math_score)

school_data_complete["average_math_score"] =school_data_complete['math_score'].mean()
school_data_complete

#Calculate the number of students with a passing math score (70 or greater)
math_score_70above =len(school_data_complete.loc[school_data_complete["math_score"]>=70])
print(math_score_70above)

# school_data_complete["math_score_70above"] =len(school_data_complete.loc[school_data_complete["math_score"]>=70])
# school_data_complete

#Calculate the percentage of students with a passing math score (70 or greater)
passing_math_per =round((math_score_70above/Total_Students )*100 ,6)
print(passing_math_per)
school_data_complete["%passing math"] =round((math_score_70above/Total_Students )*100 ,6)
school_data_complete
```

```python
# School Name : school_unique_data
# School Type: School_type
# Total Students:Total_Students
# Total School Budget:Total_budget
# Per Student Budget:Per_student_budget
# Average Math Score:Average_math_score]
# Average Reading Score: Avg_reading_score
# % Passing Math: math_score_per
# % Passing Reading:Reading_score_per
# % Overall Passing (The percentage of students that passed math and reading.):overall_pass_per
```

```python
#calculate the Reading_score average and Percentages

average_Reading_score = school_data_complete['reading_score'].mean()
print(average_Reading_score)

school_data_complete["average_Reading_score"] =school_data_complete['reading_score'].mean()
school_data_complete

#Calculate the number of students with a passing reading score (70 or greater)
Reading_score_above70 = len(school_data_complete.loc[school_data_complete["reading_score"]>=70])
print(Reading_score_above70)


#Calculate the percentage of students with a passing reading score (70 or greater)
passing_Reading_per = round((Reading_score_above70/Total_Students )*100 ,6)
print(passing_Reading_per)

school_data_complete["%passing Reading"] =round((Reading_score_above70/Total_Students )*100 ,6)
school_data_complete
```

```python
#Calculate Totalnumber of students who passed math and reading (% Overall Passing)
math_Reading_pass_students= len(school_data_complete.loc[(school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70)])
math_Reading_pass_students
```

```python
#Calculate Totalnumber of students who passed math and reading (% Overall Passing)
overall_pass_per = round((math_Reading_pass_students/Total_Students)*100 ,6)
overall_pass_per

school_data_complete["% overall pass"] =round((Reading_score_above70/Total_Students )*100 ,6)
school_data_complete
```

```python
#calculate the % passing math 


per_school_passing_math= school_data_complete[school_data_complete['math_score'] >=70]

Each_school_counts= school_data_complete["school_name"].value_counts()
Each_school_counts
Passing_per_math=per_school_passing_math.groupby(['school_name']).count()["budget"]/Each_school_counts*100
Passing_per_math
```

```python
#calculate the the % passing Reading
per_school_passing_reading= school_data_complete[school_data_complete['reading_score'] >=70]
Each_school_counts= school_data_complete["school_name"].value_counts()
Each_school_counts
Passing_per_reading=per_school_passing_reading.groupby(['school_name']).count()["budget"]/Each_school_counts*100
Passing_per_reading
```

```python
#calculate the the % Overall passing

per_school_passing_reading= school_data_complete[school_data_complete['reading_score'] >=70]
per_school_overall_passing= school_data_complete.loc[(school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70)]

Each_school_counts= school_data_complete["school_name"].value_counts()
Each_school_counts
Overall_passing_per=per_school_overall_passing.groupby(['school_name']).count()["budget"]/Each_school_counts*100
Overall_passing_per
```
