```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete
```

```python
#calculate the % passing math 
per_school_passing_math= school_data_complete[school_data_complete['math_score'] >=70]
Each_school_counts= school_data_complete["school_name"].value_counts()
Passing_per_math=per_school_passing_math.groupby(['school_name']).count()["budget"]/Each_school_counts*100
Passing_per_math
print(Passing_per_math)

```

```python
#calculate the the % passing Reading
per_school_passing_reading= school_data_complete[school_data_complete['reading_score'] >=70]
Each_school_counts= school_data_complete["school_name"].value_counts()
Each_school_counts
Passing_per_reading=per_school_passing_reading.groupby(['school_name']).count()["budget"]/Each_school_counts*100
Passing_per_reading
```

```python
#calculate the the % Overall passing

per_school_passing_reading= school_data_complete[school_data_complete['reading_score'] >=70]
per_school_overall_passing= school_data_complete.loc[(school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70)]

Each_school_counts= school_data_complete["school_name"].value_counts()
Each_school_counts
Overall_passing_per=per_school_overall_passing.groupby(['school_name']).count()["budget"]/Each_school_counts*100

school_data_complete["Per_student_budget"] = school_data_complete['budget'] / school_data_complete['size']
school_data_complete
```

```python
#Per_student_budget
school_data_complete["Per_student_budget1"] = school_data_complete['budget'] / school_data_complete['size']
school_data_complete
```

```python
school_types = school_data.set_index(["school_name"])["type"]
Grouped_school_summary=school_data_complete.groupby(['school_name']).agg(
                                                   Total_students =('Student ID','count'),
                                                    avg_math_score=('math_score', 'mean'),
                                                    avg_reading_score =('reading_score', 'mean'),
                                                  Total_Budget = ('Per_student_budget','sum') ,
                                                     per_stu_budget =('Per_student_budget1','mean')                  
                                                      ) 
                                                
                                              


Grouped_school_summary["School_Type"] = school_types   
Grouped_school_summary
```

```python
# Use Map to format all the columns
Grouped_school_summary ["Total_Budget"]=Grouped_school_summary ["Total_Budget"].map("${:,}".format)
Grouped_school_summary ["per_stu_budget"]=Grouped_school_summary ["per_stu_budget"].map("${:,}".format)

Grouped_school_summary
```

```python
# Creating dictionary and data frame      
School_summary ={"% Passing Math":Passing_per_math,"% Passing Reading":Passing_per_reading,
                     "% Overall Passing":Overall_passing_per}
School_summary
school_summary_df=pd.DataFrame(School_summary)
school_summary_df
```

```python
school_data_summary = pd.merge(Grouped_school_summary,school_summary_df,left_index=True, right_index=True)
school_data_summary
# merge_df.drop("school_name", axis =1,inplace= True)
# merge_df
```

```python
# Create a GroupBy object based upon "spending_range_per_student"

school_scores_by_Type=school_data_summary.groupby(['School_Type'])


# Get the average of each column within the GroupBy object
school_scores_by_Type[["avg_math_score", "avg_reading_score","% Passing Math","% Passing Reading","% Overall Passing"]].mean()
```
