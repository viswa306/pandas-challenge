```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# School Name : school_unique_data
# School Type: School_type
# Total Students:Total_Students
# Total School Budget:Total_budget
# Per Student Budget:Per_student_budget
# Average Math Score:Average_math_score]
# Average Reading Score: Avg_reading_score
# % Passing Math: math_score_per
# % Passing Reading:Reading_score_per
# % Overall Passing (The percentage of students that passed math and reading.):overall_pass_per
```

```python
school_data_complete
```

```python
#calculate the % passing math 
per_school_passing_math= school_data_complete[school_data_complete['math_score'] >=70]
Each_school_counts= school_data_complete["school_name"].value_counts()
Passing_per_math=per_school_passing_math.groupby(['school_name']).count()["budget"]/Each_school_counts*100
Passing_per_math
print(Passing_per_math)
#School_summary["% Passing Math"] = Passing_per_math
```

```python
#calculate the the % passing Reading
per_school_passing_reading= school_data_complete[school_data_complete['reading_score'] >=70]
Each_school_counts= school_data_complete["school_name"].value_counts()
Each_school_counts
Passing_per_reading=per_school_passing_reading.groupby(['school_name']).count()["budget"]/Each_school_counts*100
Passing_per_reading
#School_summary["% Passing Reading"] = Passing_per_reading
```

```python
#calculate the the % Overall passing

per_school_passing_reading= school_data_complete[school_data_complete['reading_score'] >=70]
per_school_overall_passing= school_data_complete.loc[(school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70)]

Each_school_counts= school_data_complete["school_name"].value_counts()
Each_school_counts
Overall_passing_per=per_school_overall_passing.groupby(['school_name']).count()["budget"]/Each_school_counts*100
#School_summary["% Overall Passing"] = Overall_passing_per

#school_summary_df =pd.DataFrame(School_summary)   
#school_summary_df 
#Per_student_budget
#per_stu_budget = school_data_complete['budget'] / school_data_complete['size']
school_data_complete["Per_student_budget"] = school_data_complete['budget'] / school_data_complete['size']
school_data_complete
```

```python
#Per_student_budget
school_data_complete["Per_student_budget1"] = school_data_complete['budget'] / school_data_complete['size']
school_data_complete
```

```python

Grouped_school_summary=school_data_complete.groupby(['school_name']).agg( Schooltype =('type','unique'),
                                                   Total_students =('Student ID','count'),
                                                    avg_math_score=('math_score', 'mean'),
                                                    avg_reading_score =('reading_score', 'mean'),
                                                  Total_Budget = ('Per_student_budget','sum') ,
                                                     per_stu_budget =('Per_student_budget1','mean')                  
                                                      ) 
                                                
                                              

Grouped_school_summary
    
```

```python
# Creating dictionary and data frame      
School_summary ={"% Passing Math":Passing_per_math,"% Passing Reading":Passing_per_reading,
                     "% Overall Passing":Overall_passing_per}
School_summary
school_summary_df=pd.DataFrame(School_summary)
school_summary_df
```

```python
school_data_summary = pd.merge(Grouped_school_summary,school_summary_df,left_index=True, right_index=True)
school_data_summary
# merge_df.drop("school_name", axis =1,inplace= True)
# merge_df
```

```python
#Bin labels must be one fewer than the number of bin edges
bins = [0,1000,2000,5000]

group_names = [ "<1000", "medium(1000-2000)", "large(2000-5000)"]
```

```python
 #Figure out the minimum and maximum budget for the school 
print(school_data_summary["Total_students"].max())
print(school_data_summary["Total_students"].min())
```

```python
school_data_summary["School_score_size"] = pd.cut(school_data_summary["Total_students"], bins, labels=group_names)
school_data_summary
```

```python
# Create a GroupBy object based upon "spending_range_per_student"

school_data_summary_group =school_data_summary.groupby("School_score_size")

# Find how many rows fall into each bin
print(school_data_summary_group["Total_students"].count())

# Get the average of each column within the GroupBy object
school_data_summary_group[["avg_math_score", "avg_reading_score","% Passing Math","% Passing Reading","% Overall Passing"]].mean()
```
