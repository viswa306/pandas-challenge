```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# ### Reading Scores by Grade(9thgrade,10th,11th,12th)
nith_graders = school_data_complete[school_data_complete["grade"]== "9th"]
nith_graders
tenth_graders = school_data_complete[school_data_complete["grade"]== "10th"]
tenth_graders
eleveth_graders = school_data_complete[school_data_complete["grade"]== "11th"]
eleveth_graders 
twelth_graders = school_data_complete[school_data_complete["grade"]== "12th"]
twelth_graders 
```

```python
#### Reading Scores by Grade(9thgrade,10th,11th,12th)
N_9th=nith_graders.groupby('school_name')["math_score"].mean()

T_10th=tenth_graders.groupby('school_name')["math_score"].mean()

e_11th=eleveth_graders.groupby('school_name')["math_score"].mean()

t_12th=twelth_graders.groupby('school_name')["math_score"].mean()

```

```python
#creating dictionary 

Math_scores_by_Grade = {"9th":N_9th,"10th":T_10th,"11th":e_11th,"12th":t_12th}
Math_scores_by_Grade
pd.DataFrame(Math_scores_by_Grade)
```
